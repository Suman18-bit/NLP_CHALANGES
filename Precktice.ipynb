```python
from transformers import AutoModelForCausalLM, AutoTokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
prompt ="Write about Virat Kohli"
input_text = prompt

generated_text = model.generate(
    tokenizer.encode(input_text, return_tensors="pt"),
    max_length=25,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    num_return_sequences=1,
    pad_token_id=tokenizer.eos_token_id,
)
print(tokenizer.decode(generated_text[0], skip_special_tokens=True))
```

```python
train_texts = tokenizer.decode(generated_text[0], skip_special_tokens=True)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

train_encodings = tokenizer(train_texts, truncation=True, padding=True)

print(train_encodings.keys())
```

```python
medical_texts = [
    "Patient presents with a history of chest pain and shortness of breath.",
    "MRI scan reveals a tumor in the frontal lobe of the brain.",
    "Blood test results indicate elevated levels of cholesterol and triglycerides.",
    "The patient was diagnosed with type 2 diabetes mellitus.",
]

tokenized_medical_texts = [tokenizer.encode(text) for text in medical_texts]

attention_masks = []
for tokenized_text in tokenized_medical_texts:
    attention_mask = [1] * len(tokenized_text)
    attention_masks.append(attention_mask)

for i, attention_mask in enumerate(attention_masks):
    print(f"Attention mask for medical text {i+1}: {attention_mask}")
```

```python
def retriever(input_ids, **kwargs):
    return {'context_input_ids': input_ids}
retriever_args = {}
generate_args = {
    "max_length": 50,
    "temperature": 0.7,
    "top_k": 50,
    "top_p": 0.95,
    "num_return_sequences": 1,
    "pad_token_id": tokenizer.eos_token_id,
}
question = "What is the latest clinical update on diabetes"
input_dict = tokenizer(question, return_tensors="pt")
results = retriever(input_dict['input_ids'], **retriever_args)
output = model.generate(inputs=results['context_input_ids'], **generate_args)

print(tokenizer.decode(output[0]))
```
