```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
bart_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
bart_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
bart_inputs = bart_tokenizer(sample_text, max_length=1024, truncation=True, return_tensors="pt")
bart_summary_ids = bart_model.generate(bart_inputs["input_ids"], num_beams=4, max_length=150, min_length=50, early_stopping=True)
bart_generated_summary = bart_tokenizer.decode(bart_summary_ids[0], skip_special_tokens=True)

pprint(bart_generated_summary)
```
