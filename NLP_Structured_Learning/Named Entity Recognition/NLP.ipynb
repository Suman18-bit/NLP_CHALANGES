```python
import pandas as pd
```

```python
import spacy
```

```python
!python -m spacy download en_core_web_sm
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
file_path = "/content/sentiment_examples.txt"
with open(file_path,'r', encoding='utf-8') as file:
  sentiment_text =file.readlines()
```

```python
token_list = []
filtered_token_list = []
pos_tag_lists = []
ner_list = []
#Creating Doc
for sentiment_txt in sentiment_text:
  doc = nlp(sentiment_txt.strip())
#Tokenization
  tokens = [token.text for token in doc]
  token_list.append(tokens)
#Removing Stop Words & Puntuations Recognization
  filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
  filtered_token_list.append(filtered_tokens)
#Parts of Speech Recognization
  pos_tags = [(token.text, token.pos_) for token in doc]
  pos_tag_lists.append(pos_tags)
#Name Entity Recognization
  ner_tags = [(ent.text, ent.label_) for ent in doc.ents]
  ner_list.append(ner_tags)
```

```python
result_df = pd.DataFrame({
    'Original Text': pd.Series(sentiment_text),
    'Tokens': pd.Series(token_list),
    'Filtered Tokens': pd.Series(filtered_token_list),
    'POS Tags': pd.Series(pos_tag_lists),
    'Name Entity': pd.Series(ner_list)
})
display(result_df)
```
