```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/sentiment labelled sentences/amazon_cells_labelled.txt',sep='\t',names = ['review','labelled_sentiment'])
df2 = pd.read_csv('/content/sentiment labelled sentences/imdb_labelled.txt',sep='\t',names = ['review','labelled_sentiment'])
df3 = pd.read_csv('/content/sentiment labelled sentences/yelp_labelled.txt',sep='\t',names = ['review','labelled_sentiment'])
```

```python
print(df1.head(),"  \n",df2.head(),"  \n",df3.head())
```

```python
df = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
```

```python
import vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
Analyzer = SentimentIntensityAnalyzer()
```

```python
df['vaderSentiment_score'] = df['review'].apply(lambda x:Analyzer.polarity_scores(x)['compound'])
```

```python
df.head()
```

```python
df['vaderSentiment'] = df['vaderSentiment_score'].map(lambda x:int(1) if x>0.05 else int(0) if x<=-0.05 else int(2))
```

```python
df.head()
```

```python
df['vaderSentiment'].unique()
```

```python
df['vaderSentiment'] = df['vaderSentiment'].map({0:'negative',1:'positive',2:'neutral'})
```
