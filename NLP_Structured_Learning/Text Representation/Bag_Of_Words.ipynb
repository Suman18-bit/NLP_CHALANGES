```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
corpus = [
    "The cat sat on the mat.",
    "Dogs bark loudly in the night.",
    "She enjoys reading books every evening.",
    "The sun rises in the east.",
    "He quickly finished his homework.",
    "Birds are flying across the sky.",
    "They played football in the park.",
    "Music makes people feel happy.",
    "The river flows silently through the valley.",
    "Children are laughing and running around."
]
```

```python
vcr = CountVectorizer()
```

```python
x = vcr.fit_transform(corpus)
```

```python
for y in x:
    print(y)
```

```python
x_arr=x.toarray()
```
