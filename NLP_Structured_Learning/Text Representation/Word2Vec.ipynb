```python
from gensim.models import Word2Vec
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
sentences = [
    "The wind carried whispers across the empty street.",
    "She painted the sky with colors no one had ever seen.",
    "Time slipped quietly through the cracks of the old clock.",
    "The river sang a song only the moon understood.",
    "He walked swiftly, chasing shadows that refused to stay still.",
    "Dreams often arrive disguised as ordinary thoughts.",
    "The stars blinked like secrets waiting to be told.",
    "Silence grew louder than any spoken word.",
    "The horizon promised change, though no one knew when.",
    "She laughed softly, as if the world had just shared a joke with her."
]
```

```python
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]

print(tokenized_sentences)
```

```python
model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=10, min_count=1, sg=0 , workers=4)
```

```python
model.wv['sky']
```

```python
model.wv['slipped']
```

```python
model.wv['street']
```

```python
model.wv.most_similar('sky')
```

```python
model.wv.most_similar('sky',topn=3) #return the top n similar
```
