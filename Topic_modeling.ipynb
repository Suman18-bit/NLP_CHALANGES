```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from pprint import pprint
```

```python
dataset = fetch_20newsgroups(subset="all",shuffle=True, random_state=32, remove=('headers', 'footers', 'quotes'))
```

```python
dataset_df = pd.DataFrame({"News":dataset.data, "Label":dataset.target})
```

```python
dataset_df.head()
```

```python
dataset_df['Label_Name']=dataset_df['Label'].apply(lambda x:dataset.target_names[x])
```

```python
dataset_df.head()
```

```python
import gensim
```

```python
from gensim.utils import tokenize
```

```python
from gensim.parsing.preprocessing import preprocess_string,strip_tags,strip_punctuation,strip_numeric,remove_stopwords,strip_short
from gensim.corpora.dictionary import Dictionary
from gensim import models
```

```python
help(preprocess_string)
```

```python
dataset_df['Clean_news'] = dataset_df['News'].apply(preprocess_string)
```

```python
dataset_df.head()
```

```python
filter = [lambda x: x.lower(),strip_tags,strip_punctuation,strip_numeric,remove_stopwords,strip_short]
dataset_df['Clean_news'] = dataset_df['News'].apply(lambda x: preprocess_string(x,filter))
```

```python
dataset_df.head()
```

```python
dataset_dictionary = Dictionary(dataset_df['Clean_news'])
```

```python
len(dataset_dictionary)
```

```python
print(dataset_dictionary.token2id)
```

```python
dataset_corpus_bow = [dataset_dictionary.doc2bow(doc) for doc in dataset_df['Clean_news']]
```

```python
len(dataset_corpus_bow)
```

```python
print(dataset_corpus_bow[1])
```

```python
tfidf = models.TfidfModel(dataset_corpus_bow)
dataset_corpus_tfidf = tfidf[dataset_corpus_bow]
```

```python
len(dataset_corpus_tfidf)
```

```python
print(dataset_corpus_tfidf[1])
```

```python
from gensim.models.ldamodel import LdaModel
```

```python
lda_bow = LdaModel(dataset_corpus_bow,num_topics=20,id2word=dataset_dictionary,passes=10)
```

```python
lda_topics_bow = lda_bow.print_topics()
for i in lda_topics_bow:
  print(i)
```

```python
lda_tfidf = LdaModel(dataset_corpus_tfidf,num_topics=20,id2word=dataset_dictionary,passes=10)
```

```python
lda_topics_tfidf = lda_tfidf.print_topics()
for i in lda_topics_tfidf:
  print(i)
```

```python
from gensim.models.lsimodel import LsiModel
```

```python
lsi_bow = LsiModel(dataset_corpus_bow,num_topics=20,id2word=dataset_dictionary)
```

```python
lsi_bow_topics = lsi_bow.print_topics()
for i in lsi_bow_topics:
  print(i)
```

```python
lsi_tfidf = LsiModel(dataset_corpus_tfidf,num_topics=20,id2word=dataset_dictionary)
```

```python
lsi_tfidf_topics = lsi_tfidf.print_topics()
for i in lsi_tfidf_topics:
  print(i)
```

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
cm_lsi_bow = CoherenceModel(model=lsi_bow,texts=dataset_df['Clean_news'],dictionary=dataset_dictionary,coherence='u_mass')
```

```python
print(cm_lsi_bow.get_coherence())
```

```python
cm_lda_bow = CoherenceModel(model=lda_bow,texts=dataset_df['Clean_news'],dictionary=dataset_dictionary,coherence='u_mass')
```

```python
print(cm_lda_bow.get_coherence())
```

```python
texts = dataset_df['Clean_news']
texts = [x for x in texts if x]
```

```python
cm_lda_bow_cv = CoherenceModel(model=lda_bow,texts=texts,dictionary=dataset_dictionary,coherence='c_v')
print(cm_lda_bow_cv.get_coherence())
```

```python
cm_lsi_bow_cv = CoherenceModel(model=lsi_bow,texts=texts,dictionary=dataset_dictionary,coherence='c_v')
print(cm_lsi_bow_cv.get_coherence())
```
