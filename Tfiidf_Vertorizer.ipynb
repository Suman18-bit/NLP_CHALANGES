```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "Clouds drift slowly across the sky.",
    "He whispered secrets into the dark night.",
    "The children danced joyfully in the rain.",
    "She completed her project with great care.",
    "Waves crashed loudly against the rocky shore.",
    "They studied diligently for the final exam.",
    "The candle flickered softly in the corner.",
    "He shouted angrily when the door slammed.",
    "Birds chirped happily at sunrise.",
    "She walked gracefully through the quiet garden."
]
```

```python
TfIdf= TfidfVectorizer()
```

```python
x=TfIdf.fit_transform(corpus)
```

```python
for i in x:
    print(i)
```

```python
print(x.toarray())
```
