```python
import nltk
```

```python
nltk.download("punkt")
```

```python
sentence = "The curious cat leapt gracefully onto the windowsill, gazing at the world beyond." # any random sentence
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
tokens = word_tokenize(sentence)
```

```python
print(tokens)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\s+', gaps=True)
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(sentence)
```

```python
print(sentences)
```

```python
from nltk.stem import PorterStemmer
```

```python
words = [
    "running", "jumping", "smiling", "whispering", "glowing", "shining",
    "wandering", "dancing", "laughing", "singing",
    "played", "walked", "smiled", "whispered",
    "glowed", "shined", "wandered", "danced", "laughed", "sang"
]
```

```python
stemer = PorterStemmer()
stems = [stemer.stem(word) for word in words]
```

```python
print(stems)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')
```

```python
LIMITIZER = WordNetLemmatizer()
lemas = [LIMITIZER.lemmatize(word, wordnet.VERB) for word in words]
```

```python
print(lemas)
```

```python
lemas_NOUN = [LIMITIZER.lemmatize(word, wordnet.NOUN) for word in words]
```

```python
print(lemas_NOUN)
```
